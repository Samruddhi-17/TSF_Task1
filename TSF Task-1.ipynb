# Task-1: Prediction using Supervised ML

#### In this task, we are going to apply Linear Regression algorithm to predict the score obtained by the students based on number of hours studied.

#### Data Source: http://bit.ly/w-data

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Reading data from given source
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data
```

```python
# Assigning the variables
# Dependent variable
y = data.iloc[:, -1].values
# Independent variable
x = data.iloc[:, :-1].values
```

```python
# Checking the correlation
data.corr()
```

From above table, we can see that there is possitive correlation between number of scores and their study hours.

```python
# Plotting Scatter plot to check the Correlation
plt.style.use("seaborn")
plt.scatter(x, y, c = "coral")
plt.title("Hours Vs Scores", fontsize = 16, fontweight = "bold")
plt.xlabel("Study Hours", fontsize = 16)
plt.ylabel("Scores", fontsize = 16);
```

From the above graph, we can see that there is a positive linear relationship between the number of hours studied and scores.

```python
# Preparing the data
# Splitting the data into training and testing sets
np.random.seed(42)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Traning the algorithm
from sklearn.linear_model import LinearRegression  
reg = LinearRegression()  
reg.fit(x_train, y_train)
```

```python
# Predicting the test results
y_pred = reg.predict(x_test)
```

```python
### Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# Plotting the regression line
plt.scatter(x_train, y_train)
y_est = reg.coef_*x_train + reg.intercept_
fig = plt.plot(x_train, y_est, linestyle = "--", lw = 3, c = "lightblue")
plt.title("Hours Vs Scores", fontsize = 16, fontweight = "bold")
plt.xlabel("Study Hours", fontsize = 16)
plt.ylabel("Scores", fontsize = 16)
plt.show()
```

```python
# Accuracy
from sklearn.metrics import r2_score
acc = r2_score(y_test, y_pred)
acc
```

```python
# Mean absolute error
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
# Predicting the score if the student studied for 9.25 hrs/day
reg.predict([[9.25]])
```

### Result
From here we can conclude that if the student studied for 9.25 hrs/day then he/she is expected to score 92.39%.
